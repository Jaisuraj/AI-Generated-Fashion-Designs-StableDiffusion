```python
pipe = StableDiffusionPipeline.from_pretrained(model_id, use_auth_token=True)
pipe = pipe.to(device)
```

```python
def make_img(desc,i):
  prompt=desc
  with autocast("cuda"):
    image = pipe(prompt, guidance_scale=7.5).images[0]
  image.save("img"+str(i)+".png")
  uploadImage("img"+str(i)+".png")
```
